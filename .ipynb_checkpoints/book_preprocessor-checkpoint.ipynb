```python
def clean_data(data):
    clean_v1 = data[~data.GenreLink.isnull()]
    clean_v2 = clean_v1.drop_duplicates(subset="BookID")
    clean_v3 = clean_v2[clean_v2["Number of raters"] > 150]
    
    publish = clean_v3.PublishDate
    year = publish.str.split().str[-1]
    year.replace(list(year[~year.str.isdigit()]), np.NaN, inplace=True)
    year.fillna(year.mode(), inplace=True)
    
    clean_v3['PublishYear'] = year
    
    clean_v3.Pages.fillna(int(clean_v3.Pages.mean()), inplace=True)
    clean_v3.Series.fillna('None', inplace=True)
    clean_v3["Book format"].fillna(clean_v3["Book format"].mode(), inplace=True)
    clean_v3.Language.fillna(clean_v3.Language.mode(), inplace=True)

    clean_v3.drop(columns="PublishDate", inplace=True)
    
    return clean_v3
```

```python
for i in range(3, 9):
    file_name = 'goodreads/books_' + str(i) + '.csv'
    
    data = pd.read_csv(file_name)
    cleaned = clean_data(data)
    
    cleaned.to_csv(file_name, index=False)
```
