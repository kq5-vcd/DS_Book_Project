```python
import pandas as pd
```

```python
books = pd.read_csv('data/book_data.csv')

books.head()
```

```python
genre = pd.read_csv('data/tags.csv')
genre.head()
```

```python
print(len(books), len(genre))
```

```python
complete_data = pd.merge(books, genre, how = 'inner', on = 'GenreLink')
len(complete_data)
```

```python
complete_data.head()
```

```python
complete_data.to_csv('data/complete_data.csv', index=False)
```
