```python
clean_v4 = clean_v3[clean_v3.PublishYear.str.isdigit()]
len(clean_v3)
```

```python
clean_v4.PublishYear = clean_v4.PublishYear.astype(int)
clean_v4 = clean_v4.drop(columns=['PublishDate'])
clean_v4.tail()
```

```python
clean_v4.Pages.fillna(int(clean_v4.Pages.mean()), inplace=True)
clean_v4.Series.fillna('None', inplace=True)
clean_v4["Book format"].fillna(clean_v4["Book format"].mode(), inplace=True)
clean_v4.Language.fillna(clean_v4.Language.mode(), inplace=True)

clean_v4.tail()
```

```python
clean_v4.to_csv('goodreads/books_1.csv', index=False)
```

```python
def clean_data(data):
    clean_v1 = data[~data.GenreLink.isnull()]
    clean_v2 = clean_v1.drop_duplicates(subset="BookID")
    
    clean_v2["Raters"] = clean_v2.loc[:, ("5 stars", "4 stars", "3 stars", "2 stars", "1 star")].apply(sum, axis=1)
    clean_v3 = clean_v2[clean_v2["Raters"] > 150]
    
    publish = clean_v3.PublishDate
    year = publish.str.split().str[-1]
    clean_v3['PublishYear'] = year
    
    clean_v4 = clean_v3[clean_v3.PublishYear.str.isdigit()]
```
